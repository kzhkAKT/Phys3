### $Q$ と $q$ を結ぶ直線上の引力と斥力

\begin{align}
F &= k \dfrac{Q \ q }{r^2}=
\begin{cases}
Qq >0 &  \text{電荷同符号は斥力}
\begin{cases}
Q>0 \ \text{and}\  q>0\\
Q<0 \ \text{and}\  q<0
\end{cases}\\[2mm]
%
Qq < 0 &  \text{電荷異符号は斥力}
\begin{cases}
Q>0 \ \text{and}\  q<0\\
Q<0 \ \text{and}\  q>0
\end{cases}
\end{cases}
\end{align}

\begin{align}
k = k_e = \dfrac{1}{4\pi \epsilon_0} \approx {\rm 8.988\times 10^9\  N⋅m^22⋅C^{-2}}
\end{align}
$\epsilon_0$ は真空の誘電率
